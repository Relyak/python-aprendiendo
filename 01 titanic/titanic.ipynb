```python
#importar las bibliotecas necesarias
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
# ignorar warnings molestos
```

```python
# cargo el conjunto de datos

datos = pd.read_csv('train.csv')

# muestro las filas del conjunto de datos

print (datos.head())
```

```python
# veo la información general de los datos

print (datos.info())
```

```python
# descripción de los datos

print (datos.describe())
```

```python
# verificar los nulos

print (datos.isnull().sum())
```

```python
# visualización

import matplotlib.pyplot as plt
import seaborn as sns

# configurar el gráfico
nulos = datos.isnull()


sns.set(style="whitegrid")
# tamaño de la figura
plt.figure(figsize=(10,6))
sns.heatmap(nulos, yticklabels=False,cbar=False,cmap='Reds')
plt.title('Valores que faltan')
plt.show()
```

```python
# Distribución de supervivientes
plt.figure(figsize=(10,6))
sns.countplot(x='Survived', data=datos)
plt.title ('Distribución de surpervivientes')
plt.show()

# Distribución de clases de pasajeros
plt.figure(figsize = (10,6))
sns.countplot(x='Pclass', data = datos)
plt.title ('Distribución de clases de pasajeros')
plt.show()

# Distribución de género
plt.figure(figsize = (10,6))
sns.countplot(x='Sex' , data = datos)
plt.title ('Distribución de sexos de pasajeros')
plt.show()
```

```python
plt.figure(figsize=(10,10))
# dropna() elimina nulos na, kde =False quitar la curva de densidad, bins = 30 intervalos de 30
sns.histplot(datos['Age'],  bins=30)
plt.title('Distribución de la edad')
plt.show()
```

```python
# Boxplot de edad por Clases

plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Age', data= datos)
plt.title ('Edad por clase de pasajero')
plt.show()
```

```python
# Supervivientes y genero
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=datos)
plt.title('Supervivencia por sexo')
plt.show()

# Relación entre supervivencia y puerto de embarque
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Embarked', data=datos)
plt.title('Supervivencia por puerto de embarque')
plt.show()


# Relación entre supervivencia y clase de pasajero
plt.figure(figsize=(10, 6))
sns.histplot(x='Survived', hue='Pclass', data=datos)
plt.title('Supervivencia por clase de pasajero')
plt.show()
```
